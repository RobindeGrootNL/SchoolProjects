```python
import matplotlib.pyplot as plt
import math
import numpy as np
import pylab as pyl
from scipy import signal
from scipy.io import wavfile
import seaborn as sns
sns.set()

%matplotlib inline
```

## Audio noise

The audio noise is worked on first. Let's import the data!

```python
def import_audio(filepath):
    try:
        fs, data = wavfile.read(filepath)
    except:
        print("ERROR: Please use a .wav file")
    
    # From stereo to mono
    try:
        if data.shape[1] > 1:
            monodata = (data[:,0] / 2 + data[:,1] / 2)
            print("File is transformed to a mono file.")
    except:
        monodata = data
        print("File is already a mono file.")
    return fs, data
```

```python
filenames = ['./Data/noise1.wav', './Data/noise2.wav', './Data/noise3.wav']

fs_1, noise_1 = import_audio(filenames[0])
fs_2, noise_2 = import_audio(filenames[1])
fs_3, noise_3 = import_audio(filenames[2])    
```

### Noise 1

Firstly, the processes are performed on noise 1.

#### Periodogram analysis

The following plot portrays the periodogram analysis of noise 1.

```python
f_p1, Pxx_den_p1 = signal.periodogram(noise_1, fs_1)
plt.semilogy(f_p1, Pxx_den_p1)
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.show()
```

```python
np.mean(Pxx_den_p1)
```

```python
f_p1, Pxx_spec_p1 = signal.periodogram(noise_1, fs_1, 'flattop', scaling='spectrum')
plt.figure()
plt.semilogy(f_p1, np.sqrt(Pxx_spec_p1))
plt.xlabel('frequency [Hz]')
plt.ylabel('Linear spectrum [V RMS]')
plt.show()
```

```python
np.sqrt(Pxx_spec_p1.max())
```

#### Welsh method

The following plots portray the Welsh method analysis of the audio noise.

```python
f_w1, Pxx_spec_w1 = signal.welch(noise_1, fs_1)
plt.semilogy(f_w1, Pxx_spec_w1)
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.show()
```

```python
np.mean(Pxx_spec_w1)
```

```python
f_w1, Pxx_spec_w1 = signal.welch(noise_1, fs_1, 'flattop', scaling='spectrum')
plt.semilogy(f_w1, Pxx_spec_w1)
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.show()
```

```python
np.sqrt(Pxx_spec_w1.max())
```

The noise spectra show a clear flat characteristic. The only colour noise that has such a flat characteristic is white noise, and therefore the noise present in this file is to be considered as white noise.
